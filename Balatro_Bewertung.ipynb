# 🃏 Pokerhände erkennen

In Poker gibt es verschiedene Kombinationen – von schwach bis stark:

| Nummer | Name | Beschreibung |
|--------|------|-------------|
| 0 | High Card | Keine Kombination |
| 1 | Pair | Zwei gleiche Werte |
| 2 | Two Pair | Zwei verschiedene Paare |
| 3 | Three of a Kind | Drei gleiche Werte |
| 4 | Straight | Fünf Karten in einer Reihe |
| 5 | Flush | Alle fünf Karten gleiche Farbe |
| 6 | Full House | Drilling + Paar |
| 7 | Four of a Kind | Vier gleiche Werte |
| 8 | Straight Flush | Straight + Flush |
| 9 | Royal Flush | Straight Flush ab 10 |

## 🎯 Unsere Aufgabe

Schreibt eine Funktion `bewertung(hand)`, die eine Hand aus 5 Karten bekommt und die richtige Nummer zurückgibt.

## 1. Schritt

Bevor wir Pokerhände erkennen können, brauchen wir ein **Kartendeck**.

👉 Wir bauen das zusammen:


```python
import random
werte = "23456789TBDKA"
farben = "♠♦♣♥"

deck = []
for farbe in farben:
    for wert in werte:
        deck.append(farbe+wert)
random.shuffle(deck)
```

* `["K","K","K","2","5"]` → `[3, 1, 1]`
