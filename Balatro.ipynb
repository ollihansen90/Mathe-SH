## 💡 Neue CSS-Eigenschaften

### 🔹 `display: flex;`

Damit wird unsere Karte zu einem **flexiblen Bereich**.

👉 Die Elemente innerhalb der Karte können dadurch leichter angeordnet werden.


### 🔹 `flex-direction: column;`

Die Elemente werden **untereinander** angeordnet.

👉 Also:

* oben → Kartenwert
* unten → Kartensymbol


### 🔹 `justify-content: space-between;`

Die Elemente bekommen möglichst viel Abstand voneinander.

👉 Dadurch steht:

* der Wert oben
* das Symbol unten


### 🔹 `font-family: Georgia, serif;`

Damit ändern wir die Schriftart der Karte.

👉 Die Karte sieht dadurch mehr wie eine echte Spielkarte aus.
