## ⚙️ Setup

Zuerst laden wir alles was wir brauchen — das kennt ihr schon vom letzten Mal!

```python
import random
import ipywidgets as widgets
from IPython.display import display

werte = "23456789TBDKA"
farben = "♠♦♣♥"

deck = []
for farbe in farben:
    for wert in werte:
        deck.append(farbe + wert)

random.shuffle(deck)

def occurrence(string):
    string = sorted(string)
    out = [1]
    for i,j in zip(string[:-1], string[1:]):
        if i==j:
            out[-1]+=1
        else:
            out.append(1)
    return sorted(out)[::-1]

def bewertung(hand):
    """0: High Card
       1: Pair
       2: Two Pair
       3: Three of a Kind
       4: Straight
       5: Flush
       6: Full House
       7: Four of a Kind
       8: Straight Flush
       9: Royal Flush
       """
    straight = False
    flush = False

    hand = sorted(hand, key=lambda x: "23456789TBDKA".find(x[1]))
    werte = [h[1] for h in hand]
    gruppen = occurrence(werte)
    if gruppen[0]==2:
        if gruppen[1]==1:
            return 1
        elif gruppen[1]==2:
            return 2
    if gruppen[0]==3:
        if gruppen[1]==1:
            return 3
        else:
            return 6
    if gruppen[0]==4:
        return 7 # Four of a Kind
    farben = list(set(h[0] for h in hand))

    if "".join(werte) in "23456789TBDKA":
        straight = True

    if len(farben)==1:
        flush = True
    if straight and flush:
        if werte[0]=="T":
            return 9 # Royal Flush
        else:
            return 8 # Straight Flush

    if flush: return 5 # Flush

    if straight: return 4 # Straight

    return 0

def spielkarte(k):
    wert = k[1]
    farbe = k[0]
    if wert=="T":
        wert = "10"

    hintergrund = {
        "♠": "rgb(30, 30, 40)",
        "♦": "rgb(40, 75, 160)",
        "♣": "rgb(20, 110, 55)",
        "♥": "rgb(150, 20, 35)"
    }
    grau = "rgb(240, 235, 225)"
    karte = f"""
    <div style='
        background-color: {hintergrund[farbe]};
        width: 50px;
        height: 70px;
        padding: 10px;
        border-radius: 10%;
        display: flex;
        flex-direction: column;
        justify-content: space-between;
        font-family: Georgia, serif;
    '>
    <div style='
        font-size: 20px;
        width: 100%;
        text-align: left;
        color: {grau};

    '>{wert}</div>
    <div style='
        font-size: 50px;
        width: 100%;
        text-align: right;
        color: {grau};
    '>{farbe}</div>
    </div>
    """
    return karte
```
