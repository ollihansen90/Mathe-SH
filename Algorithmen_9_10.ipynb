## Sieb des Eratosthenes

````python
def primlist(k):
    list = [2]
    for i in range(3, k+1):
        unteilbar = True
        for element in list:
            if i % element == 0:
                unteilbar = False
        if unteilbar:
            list.append(i)
    return list

#Farbübersicht
#     lila = '\033[95m'     blau = '\033[94m'     cyan = '\033[96m'    dunkelcyan = '\033[36m'
#     grun = '\033[92m'     gelb = '\033[93m'     rot = '\033[91m'    grau    = \033[90m
#     normal = '\033[0m'    fett = '\033[1m'        unterstrichen = '\033[4m'

def eratosthenes(len_list, stopper):    
    farben = ['\033[95m','\033[93m','\033[94m','\033[96m','\033[92m','\033[90m','\033[36m']
    primes = primlist(len_list)
    print(" ", end="")                                #Korrektur erste Zeile

    for i in range(1,len_list+1):
        if i < 10:
            print(" ", end="")                        # Zahlen in einer Reihe
        if i < 100:
            print(" ", end="")
        
        if i in primes[0:stopper]:            #bisherige Primzahlen rot und unterstrichen
            print('\033[91m' + '\033[4m', end="")
        else:                                                     
            if i in primes and primes[stopper] > len_list**0.5:     #alle Primzahlen ab Wurzel(Listenlänge) färben
                print('\033[91m' + '\033[4m', end="")
            farbe_unbekannt = True                # Nichtprimzahlen nach dem ersten Teiler aus Primzahlliste mit Farbliste färben
            farbnummer = 0
            while farbe_unbekannt and farbnummer < stopper:
                if i%primes[farbnummer]==0:
                    print(farben[farbnummer],end="")
                    farbe_unbekannt = False
                farbnummer += 1

        print(str(i), end="")                     # Ausgabe
        if i % 10 == 0 and i != 1:            # neue Zeile bei 10er-Sprung
            print("")
        print('\033[0m', end=" ")             # neue Zeile schwarz und normal

    print("\n" + '\033[91m' + '\033[4m' + "Primzahlen" + '\033[0m')     #Ausgabe der benutzen Teiler mit Farbigkeiten
    for k in range(0,stopper):
        print(farben[k] + "teilbar durch " + str(primes[k]))
    print( '\033[0m' + "-------------------------------------------------------------------------------------------")

eratosthenes(50,1)
eratosthenes(50,2)
eratosthenes(50,3)
eratosthenes(50,5)
eratosthenes(300,7)

print(primlist(10000))
````

# (Moderner) euklidischer Algorithmus

$a = q_1 \cdot r_0 + r_1 \; \text{mit} \; b = r_0$

$r_0 = q_2 \cdot r_1 + r_2$

$r_1 = q_3 \cdot r_2 + r_3$

$\vdots$

$r_{n-1} = q_{n+1} \cdot r_n + 0$

$ggT(a,b) = r_n$

Für die Implementierung in Python brauchen wir den sogenannten **Modulo-Operator ```%```**, der den Rest der Division angibt.
Folglich gilt $$6 \% 3 = 0 \quad\text{und}\quad 5 \% 3 = 2.$$


Außerdem hilft der **Operator ```//```** bei der Ausgabe der einzelnen Schritte, da er das Ergebnis der Division auf die nächstkleinere ganze Zahl rundet. Es gilt also $$6 \, // \, 3 = 2 \quad\text{und}\quad 5 \, // \, 3 = 1.$$

````python
# rekursiv
def ggT_rec(a, b):
    print("a =", a, "b =", b, "\n")

    if b == 0:                          # wenn b gleich Null
        print("ggT =", a)               # Ausgabe auf der Konsole
        return a                        # Rückgabe, die einer Variablen zugeordnet werden kann
    
    else:                               # sonst, d.h. wenn b gleich Null
        print(a, "=", a // b, "*", b, "+", a % b)
        return ggT_rec(b, a % b)


# iterativ
def ggT_it(a, b):
    print("a =", a, "b =", b, "\n")

    while b != 0:                       # solange b ungleich Null
        temp = a % b                    # Hilfsvariable
        print(a, "=", a // b, "*", b, "+", temp)
        a = b
        b = temp
    print("ggT =", a)                   # Ausgabe auf der Konsole
    return a                            # Rückgabe, die einer Variablen zugeordnet werden kann

#ggT_it(1078, 139) # Testaufruf mit a = 1078, b = 139
````
