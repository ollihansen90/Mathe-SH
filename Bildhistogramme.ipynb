```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io

image_url = "https://www.nationalpark-bayerischer-wald.bayern.de/pic/virtuell/pressebilder/11/2443867_o.jpg"
img = io.imread(image_url)[::4,::4,:]/255 # Nimm nur jeden vierten Pixel, damit das Bild nicht so groß ist

plt.figure()
plt.imshow(img)
plt.show()
```

```python
img_sw = img@np.array([0.299, 0.7152, 0.0722]) # Bild zu Schwarz/Weiß; @ ist das selbe wie np.matmul(...)
print(img_sw.shape)

plt.figure()
plt.imshow(img_sw) # Plotte Schwarz/Weiß-Bild
plt.show()

plt.figure()
plt.hist(img_sw.flatten(), bins=256) # Plotte Histogramm
plt.show()
```
