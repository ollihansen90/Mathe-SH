```python
import random
import matplotlib.pyplot as plt

def initialisiere(n_zellen):
    output = []
    for i in range(n_zellen):
        output.append(random.randint(0,1))
    return output

def update(welt):
    output = []
    for i in range(len(welt)):
        nachbarn = welt[i-2]+welt[i-1]+welt[i]+welt[(i+1)%len(welt)]+welt[(i+2)%len(welt)]
        if nachbarn==3 and random.random()<0.1:
            output.append(1)
        elif nachbarn==2 or nachbarn==4:
            output.append(1)
        else:
            output.append(0)
    return output

random.seed("Olli")
schritte = 300
weltgroesse = 500
welten = []
welten.append(initialisiere(weltgroesse))
for i in range(schritte):
    welten.append(update(welten[-1]))
plt.figure()
plt.imshow(welten[::-1])
plt.axis("off")
plt.tight_layout()
plt.savefig("bild.png")
plt.show()
```
