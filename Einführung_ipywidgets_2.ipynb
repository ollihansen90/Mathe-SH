## 🧩 Aufgabe: Unser Haiku mit CSS gestalten

Jetzt gestalten wir unser Haiku mit **CSS**.
CSS sorgt dafür, dass Texte und Elemente **schöner aussehen**.

Wir schreiben die CSS-Befehle direkt in das `style`-Attribut von `<p>`.

---

### 🎨 Wichtige CSS-Eigenschaften

* `color:`
  bestimmt die **Schriftfarbe**
  Beispiel: `color: yellow;`

* `background-color:`
  bestimmt die **Hintergrundfarbe**
  Beispiel: `background-color: grey;`

* `font-size:`
  bestimmt die **Schriftgröße**
  Beispiel: `font-size: 40pt;`

* `font-weight:`
  bestimmt die **Schriftdicke**
  Beispiel: `font-weight: bold;`

* `text-align:`
  bestimmt die **Ausrichtung des Textes**
  Beispiele:

  * `left` = links
  * `center` = mittig
  * `right` = rechts

* `width:`
  bestimmt die **Breite** des Textfeldes
  Beispiel: `width: 50%;`

* `padding:`
  bestimmt den **Innenabstand** zwischen Text und Rand
  Beispiel: `padding: 10pt;`

* `border:`
  erstellt einen **Rand** um das Element
  Beispiel: `border: 1pt solid black;`

* `border-radius:`
  macht die **Ecken rund**
  Beispiel: `border-radius: 2pt;`
