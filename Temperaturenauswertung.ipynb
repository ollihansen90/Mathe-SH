```python
import numpy as np
import matplotlib.pyplot as plt

raw = np.genfromtxt("temperaturen.txt", delimiter=";")
sprung = raw[1:, 0] - raw[:-1,0]

datalist = []
i_alt = 0
for i in range(len(sprung)):
    if sprung[i]<0:
        datalist.append(raw[i_alt+1:i])
        i_alt = i
datalist.append(raw[-i_alt+1:])

for data in datalist:
    if len(data)>1:
        plt.figure(figsize=[10, 5])
        plt.title("Raspberry Pi Pico - Temperaturverlauf")
        plt.plot((data[:,0]-data[0,0])/(60**2), data[:, 1])
        plt.grid()
        plt.xlabel("Stunden")
        plt.ylabel("Temperatur in °C")
        plt.show()
```
