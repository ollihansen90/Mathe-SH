## 🔎 Wie funktioniert unser Spielfeld?

Unser Spielfeld ist eine **Liste von Listen** (also eine Tabelle 📋).  

Jedes Feld enthält eine Zahl, die eine bestimmte Bedeutung hat:

- `0` → Wasser 🌊  
- `1` → Schiff 🚢  
- (Weitere Zustände kommen später 😉)

Beim Zeichnen wird jede Zahl automatisch in ein passendes Symbol umgewandelt.  
So sieht unser Spiel direkt übersichtlich aus!

---

## 🚀 Was ist heute unser Ziel?

Heute machen wir unser Spiel **schlauer** 🧠✨  

Bisher werden Schiffe einfach gesetzt — selbst wenn:

❌ sie aus dem Spielfeld herausragen  
❌ sie sich überschneiden  
❌ sie vielleicht direkt aneinanderstoßen  

Das ändern wir jetzt!

🎯 **Ziel der Stunde:**  
Wir schreiben eine Funktion, die prüft:

> 👉 *Darf dieses Schiff hier wirklich gesetzt werden?*

Erst wenn alles gültig ist, wird das Schiff platziert ✅  
Sonst geben wir eine passende Rückmeldung ❌



```python
def reset_spielfeld():
    return [[0 for i in range(10)] for j in range(10)]

def male_feld(feld):
    print("  │ 1 2 3 4 5 6 7 8 9 0 │  ")
    print("──┼─────────────────────┼─")
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    for i, zeile in enumerate(feld):
        string = alphabet[i] + " │ "
        for zeichen in zeile:
            string = string+"~sxtz"[zeichen] + " "
        string = string + "│"
        print(string)
    print("──┼─────────────────────┼─")


def setze_schiff(feld, pos, dir, l):
    # pos: Koordinaten
    # dir: 0 horizontal (waagerecht), 1 vertikal (senkrecht)
    # l: Länge
    if (dir and pos[0]+l>10) or (not dir and pos[1]+l>10):
        return feld
    for i in range(l):
        if dir==0:
            feld[pos[0]][pos[1]+i] = 1
        else:
            feld[pos[0]+i][pos[1]] = 1

    return feld


spielfeld = reset_spielfeld()

spielfeld = setze_schiff(spielfeld, [2,2], 1, 4)
spielfeld = setze_schiff(spielfeld, [7,1], 0, 3)
spielfeld = setze_schiff(spielfeld, [4,1], 0, 3)
male_feld(spielfeld)
```

## 🧩 Darf das Schiff hier stehen? - Unsere neue Funktion `erlaubt`

Bisher werden Schiffe einfach gesetzt 🚢  
Aber ein richtiges Spiel braucht Regeln!

Bevor wir ein Schiff platzieren, wollen wir prüfen:

🔎 Ist genug Platz auf dem Spielfeld?  
🔎 Bleibt das Schiff innerhalb der Grenzen?  
🔎 Überschneidet es sich nicht mit einem anderen Schiff?  

Genau dafür schreiben wir jetzt eine neue Funktion:

### 🛠️ Die Funktion `erlaubt(...)`

Unsere Funktion soll folgende Informationen bekommen:

- `feld` → das aktuelle Spielfeld 🌊  
- `pos` → die Startposition als Liste, z.B. `[Spalte, Zeile]` 📍  
- `dir` → die Ausrichtung  
  - `0` = waagerecht ➡️  
  - `1` = senkrecht ⬇️  
- `l` → die Länge des Schiffes 🚢  

Die Funktion soll **nicht** das Schiff setzen!  
Sie soll nur eine Frage beantworten:

> 👉 Darf das Schiff hier gesetzt werden?

### 🎯 Rückgabewert

- `True` ✅ → Das Schiff darf gesetzt werden  
- `False` ❌ → Das Schiff darf NICHT gesetzt werden  

So trennen wir klar:

🧠 `erlaubt` = prüfen  
🚢 `setze_schiff` = wirklich setzen  

Das macht unseren Code übersichtlicher und professioneller ✨


```python
def padding(feld):
    output = [[0 for i in range(12)] for j in range(12)]
    for i in range(10):
        for j in range(10):
            output[i+1][j+1] = feld[i][j]
    return output

def erlaubt(feld, pos, dir, l):
    hilfsfeld = [[1 for i in range(10)] for j in range(10)]
    p_feld = padding(feld)
    # Rand
    if dir==0:
        for zeile in range(10):
            for spalte in range(10-l+1, 10):
                hilfsfeld[zeile][spalte] = 0
    else:
        for zeile in range(10-l+1, 10):
            for spalte in range(10):
                hilfsfeld[zeile][spalte] = 0
    # Schiff-Nachbarn
    for zeile in range(10):
        for spalte in range(10):
            if hilfsfeld[zeile][spalte] == 0:
                continue
            summe = 0
            summe = summe + p_feld[zeile+1+1][spalte+1] # drunter
            summe = summe + p_feld[zeile+1+1][spalte+1+1] # unten rechts
            summe = summe + p_feld[zeile+1][spalte+1+1] # rechts
            summe = summe + p_feld[zeile+1-1][spalte+1] # drüber
            summe = summe + p_feld[zeile+1-1][spalte+1-1] # oben links
            summe = summe + p_feld[zeile+1][spalte+1-1] # links
            summe = summe + p_feld[zeile+1+1][spalte+1-1] # unten links
            summe = summe + p_feld[zeile+1-1][spalte+1+1] # oben rechts
            if summe>0:
                hilfsfeld[zeile][spalte] = 0

    # Crash
    for zeile in range(10):
        for spalte in range(10):
            if hilfsfeld[zeile][spalte] == 0:
                continue
            summe = 0
            for i in range(l):
                if dir==0:
                    summe += p_feld[zeile+1][spalte+1+i]
                    summe += p_feld[zeile+1-1][spalte+1+i]
                    summe += p_feld[zeile+1+1][spalte+1+i]
                    summe += p_feld[zeile+1][spalte+1+i+1]
                else:
                    summe += p_feld[zeile+1+i][spalte+1]
                    summe += p_feld[zeile+1+i][spalte+1-1]
                    summe += p_feld[zeile+1+i][spalte+1+1]
                    summe += p_feld[zeile+1+i+1][spalte+1]
            if summe>0:
                hilfsfeld[zeile][spalte] = 0

    male_feld(hilfsfeld)
    if hilfsfeld[pos[0]][pos[1]]==1:
        return True
    else:
        return False

spielfeld = reset_spielfeld()
setze_schiff(spielfeld, [4,5], 0, 4)
male_feld(spielfeld)

print(erlaubt(spielfeld, [0,7], 1, 4))

# TODO: Diagonalcrash
```

## 🚢 Unsere Flotte - Die internationale Standard-Version

Jetzt legen wir fest, **welche Schiffe** in unserem Spiel vorkommen sollen.

Wir spielen mit der klassischen internationalen Standardflotte:

- 1 × Schiff der Länge **5**
- 1 × Schiff der Länge **4**
- 2 × Schiffe der Länge **3**
- 1 × Schiff der Länge **2**

Das bedeutet:

👉 Insgesamt **5 Schiffe**  
👉 Insgesamt **17 belegte Felder**

---

### 🎯 Was heißt das für unser Programm?

Wir werden die Schiffe **nacheinander setzen**.

Für jedes Schiff:

1️⃣ Fragen wir mit `input()`:
- Wo soll das Schiff starten? 📍
- In welche Richtung? (0 = waagerecht ➡️, 1 = senkrecht ⬇️)

2️⃣ Prüfen wir mit unserer Funktion `erlaubt(...)`:
- Ist die Position gültig? ✅ oder ❌

3️⃣ Falls erlaubt:
- Wird das Schiff gesetzt 🚢

4️⃣ Falls nicht erlaubt:
- Muss der/die Spieler*in es erneut versuchen 🔁

---

### 🧠 Warum setzen wir die Schiffe nacheinander?

So behalten wir die Kontrolle:

- Wir wissen genau, welches Schiff gerade gesetzt wird.
- Wir können eine klare Struktur im Code behalten.
- Unser Programm wird übersichtlich und logisch aufgebaut.
