# üß† Was bedeutet if __ name __ == "__ main __":

Dieser spezielle Code-Block wird in Python verwendet, um zu bestimmen, wann ein Programm wirklich gestartet werden soll.

In Python hat jede Datei einen Namen.
Wenn du eine Datei direkt startest (z. B. mit ‚ÄûRun‚Äú oder √ºber die Konsole),
dann bekommt sie automatisch den Namen:

```

```
Das bedeutet:


```

    main()
```
wird nur dann ausgef√ºhrt, wenn die Datei direkt gestartet wird.

## üß© Warum ist das wichtig?

Weil wir so festlegen:

üëâ ‚ÄûStarte das Spiel nur, wenn das Programm direkt ausgef√ºhrt wird.‚Äú

Das hei√üt:

Wenn du die Datei direkt startest ‚Üí main() wird ausgef√ºhrt
‚Üí das Spiel beginnt

Wenn die Datei nur importiert wird (z. B. in einem anderen Programm)
‚Üí main() wird nicht automatisch gestartet
‚Üí das ist wichtig, um Fehler zu vermeiden
