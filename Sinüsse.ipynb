#Sinus

$f(x) = \text{sin}(1\cdot x)+\text{sin}(2\cdot x)+\text{sin}(3\cdot x)+\text{sin}(4\cdot x)+\ldots$

$g(x) = \text{sin}(x)+\frac{1}{3}\cdot\text{sin}(3\cdot x)+\frac{1}{5}\cdot\text{sin}(5\cdot x)+\frac{1}{7}\cdot\text{sin}(7\cdot x)+\ldots$

$h(x) = \text{sin}(x)-\frac{1}{2}\cdot\text{sin}(2\cdot x)+\frac{1}{3}\cdot\text{sin}(3\cdot x)-\frac{1}{4}\cdot\text{sin}(4\cdot x)+\ldots$

$i(x) = \text{cos}(x)+\frac{1}{9}\cdot\text{cos}(3\cdot x)+\frac{1}{25}\cdot\text{cos}(5\cdot x)+\frac{1}{49}\cdot\text{cos}(7\cdot x)+\ldots$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500
x = np.linspace(-2*np.pi, 2*np.pi, num=1000)
y = np.sin(x)
f = lambda x: np.sum(np.sin(np.arange(N)[:,None]*x), axis=0)
g = lambda x: np.sum(1/(2*np.arange(N)+1)[:,None]*np.sin((2*np.arange(N)+1)[:,None]*x), axis=0)
h = lambda x: np.sum((-1)**np.arange(N)[:,None]*np.sin((np.arange(N)[:,None]+1)*x)/(np.arange(N)[:,None]+1), axis=0)
i = lambda x: np.sum(1/(2*np.arange(N)+1)[:,None]**2*np.cos((2*np.arange(N)+1)[:,None]*x), axis=0)

plt.figure(figsize=[10,4])
plt.plot(x,f(x))
plt.grid()
plt.title("f(x)")
plt.show()
```

```python
plt.figure(figsize=[10,4])
plt.plot(x,f(x))
plt.grid()
plt.title("f(x)")
plt.show()
plt.figure(figsize=[10,4])
plt.plot(x,g(x))
plt.grid()
plt.title("g(x)")
plt.show()
plt.figure(figsize=[10,4])
plt.plot(x,h(x))
plt.grid()
plt.title("h(x)")
plt.show()
plt.figure(figsize=[10,4])
plt.plot(x,i(x))
plt.grid()
plt.title("i(x)")
plt.show()
```
