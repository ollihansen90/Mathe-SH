## üß© Wortketten: von Wort zu Wort mit nur einem Buchstaben (aus letzer Woche)

In dieser Aufgabe besch√§ftigen wir uns mit sogenannten **Wortketten** (auch *Wortleitern* genannt).

Eine Wortkette ist eine Abfolge von W√∂rtern, bei der sich **von einem Wort zum n√§chsten immer genau ein Buchstabe √§ndert**.  
Alle W√∂rter m√ºssen dabei g√ºltige W√∂rter sein.

### Beispiel
Hund ‚Üí Wund ‚Üí Wand ‚Üí Wald ‚Üí bald ‚Üí Ball ‚Üí Fall ‚Üí Fell ‚Üí Feld

Zwischen jeweils zwei aufeinanderfolgenden W√∂rtern unterscheidet sich **nur ein Buchstabe**.

### Ziel dieses Notebooks

In diesem Notebook werden wir Schritt f√ºr Schritt lernen,

- wie man W√∂rter in Python speichert und vergleicht,
- wie man √ºberpr√ºft, ob sich zwei W√∂rter nur in einem Buchstaben unterscheiden,
- und wie man mit Python selbst solche Wortketten findet oder erstellt.

Die folgenden Unterlagen dienen als Vorkenntnisse zur Bearbeitung dieses Notebooks:

```python
from tqdm.auto import tqdm


def distanz(w1, w2):
    w1 = w1.upper()
    w2 = w2.upper()
    ausgabe = 0
    # Hier eigenen Code einbauen
    # Mund, Mond
    # M     M       Passt
    #  u     o      Passt nicht -> ausgabe = ausgabe+1
    #   n     n     Passt
    #    d     d    Passt
    for i in range(len(w1)):
        if w1[i]!=w2[i]:
            ausgabe = ausgabe+1


    return ausgabe


def levenshtein(w1, w2):
    w1 = w1.upper()
    w2 = w2.upper()
    if len(w1)==len(w2): return distanz(w1, w2)
    tabelle = [[0 for _ in range(len(w1)+1)] for _ in range(len(w2)+1)]
    for i in range(len(w1)+1):
        tabelle[0][i] = i
    for i in range(len(w2)+1):
        tabelle[i][0] = i


    for i in range(1, len(w1)+1):
        for j in range(1, len(w2)+1):
            oben = tabelle[j-1][i] + 1 # L√ºcke einf√ºgen
            links = tabelle[j][i-1] + 1 # L√ºcke einf√ºgen
            if w1[i-1]==w2[j-1]: # Match (gleiche Buchstaben)
                diag = tabelle[j-1][i-1] + 0
            else: # Mismatch (unterschiedliche Buchstaben)
                diag = tabelle[j-1][i-1] + 1
            tabelle[j][i] = min([oben, links, diag])
    return tabelle[-1][-1]


def generate_gruppen(wortliste):
    gruppenliste = [[wortliste[0]]]
    for wort in tqdm(wortliste[1:]):
        enthalten = []
        for idx in range(len(gruppenliste)):
            gruppe = gruppenliste[idx]
            for gruppenwort in gruppe:
                if levenshtein(wort, gruppenwort)==1:
                    gruppe.append(wort)
                    enthalten.append(idx)
                    break
        if len(enthalten)==0:
            gruppenliste.append([wort])
        else:
            neueliste = []
            for idx in enthalten[::-1]:
                neueliste = neueliste + gruppenliste[idx]
                gruppenliste.pop(idx)
            neueliste = list(set(neueliste))
            gruppenliste.append(neueliste)
    return gruppenliste
```
