# Schleifen

In dem heutigen Teil sollen euch Schleifen nähergebracht werden. Wir benötigen Schleifen, wenn wir eine Aktion für meherer Dinge ausführen wollen oder so lange, bis wir etwas erreicht haben. Dabei gibt es folgende Schleifen-Typen:
* Wiederhol-Schleifen oder auch ```while```-Schleifen
* Zähl-Schleifen oder auch ```for```-Schleifen

Als erstes wollen wir uns auf die while-Schleifen fokussieren.

## ```while```-Schleifen
Ihr habt letzte Woche bereits ```if```-Abfragen kennengelernt, die eine Aktion ausführen, **falls** eine bestimmte Bedingung ```True``` ist.

Wir nutzen die ```while```-Schleife, wenn wir etwas **solange** ausführen wollen, wie eine bestimmte Bedingung ```True``` ist.

>**Solange** *Bedingung* gilt**:**
>> tue *Anweisung*. 

Die *Bedingung* ist hierbei sehr oft ein Counter oder ähnliches. Damit wir unsere *Anweisung* nicht ewig wiederholen, schließen wir in der Regel eine Erhöhung des Counters in unserer *Anweisung* ein.

Es folgen zwei Beispiele:

````python
print("Beispiel 1")
counter = 0
while counter<2:
    print(counter)
    counter = counter+1

print("Beispiel 2")
counter = 2
while 13%counter>0:
    print(counter)
    counter = counter+1
````

## Endlos-Schleifen

Leider erzeugen einige, häufig ungünstig programmierte Schleifen keinen normalen Ausgabewert, sondern unendlich viele Ausgabewerte, ohne je eine endgültige Ausgabe zu liefern ("zu terminieren").

Dies nennt man eine Endlosschleife.

Eine Endlosschleife terminiert nicht, weil z. B. die ```solange```-Bedingung einer ```while```-Schleife dauerhaft erfüllt ist oder bei einer ```if```-Schleife die Abbruchbedingung niemals erfüllt wird.

Solltet ihr in einer Endlosschleife landen, könnt ihr den Algorithmus "per Hand" beenden, indem ihr auf das Stopp-Symbol geht oder STRG+C drückt.

### Beispiel einer Endlosschleife
Im folgenden Beispiel haben zwei Schwestern eine Tüte Bonbons gekauft.Sie beschließen, dass sie beide je einen Bonbon essen, solange sie diese unter sich aufteilen können.

Ihrem kleinen Bruder versprechen sie, dass er alle aus der Tüte erhält, sobald sie die Tüte nicht aufteilen können und versprechen ihm, solang neue Tüten kaufen und Bonbons zu essen, bis er etwas abbekommt.

Ungünstigerweise enthält jede Tüte 10 Bonbons.
Die Ausgabe gibt an, wie viele Bonbons noch in der aktuell gefüllten Tüte sind.

Vollziehe nach, was im Code geschieht.

````python
bonbons=10
while bonbons%2==0:
    bonbons=bonbons-2
    if bonbons<0:
        bonbons=bonbons+10
    print(bonbons)
print("Bruder erhält", bonbons, "Bonbons")
````

## Listen

Durch die Wertzuweisung ```liste[i] = neues_element``` wird das i-te Element direkt verändert.

````python
# So erstellt man eine leere Liste
liste1 = []
# identisch zu
liste1 = list()

# und so eine bereits befüllte Liste
prog_sprachen = ["Python", "C++", "Java", "R"]

# Zugriff auf das vierte Element
print(prog_sprachen[3])
# identisch zu
print(prog_sprachen[-1])

# Setzen das dritte Element auf "PHP"
prog_sprachen[2] = "PHP"
print(prog_sprachen)
````

### Hilfreiche Funktionen bzw. Methoden

Es gibt einige Funktionen, die die Arbeit mit Listen vereinfachen:
* ```len(liste)``` gibt die Länge der Liste, d.h. die Anzahl der Elemente, aus.
* ```sorted(liste)``` sortiert die übergebene Liste aufsteigend bzw. alphabetisch.
* ```sum(liste)``` gibt die Summe und ```max(liste)``` das Maximum einer Liste aus Zahlen zurück.
* ```liste.append(element)``` fügt das übergebene Element an das Listenende an.
* ```liste.pop()``` entfernt das letzte Element und gibt es aus.
* ```liste.index(element)``` findet das übergebene Element und gibt die entsprechende Position aus.

````python
temperaturen = [10,8,16,11,6,5]

# Berechnung der durchschnittlichen Temperatur mittels sum und len
durchschnitt_temp = sum(temperaturen)/len(temperaturen)
print(durchschnitt_temp)

# Entfernen der letzten Temperatur
temperaturen.pop()

# sorted verändert temperaturen nicht!
print(sorted(temperaturen)) 
print(temperaturen)
````

## ```for```-Schleifen

```for```-Schleifen sind, wie der Name bereits sagt, ebenfalls Schleifen, die jedoch immer endlich sind. Bei Python funktionieren Schleifen so, dass sie ein iterierbares Objekt bekommen (z. B. eine Liste) und nacheinander die Einträge des Objektes durchläuft.

````python
liste = [1,2,3, "hallo", 4, 98920]
for eintrag in liste:
    print(eintrag)
````

For-Schleifen sind in der Syntax ähnlich wie While-Schleifen, die Abbruchbedingung ist jedoch immer so gewählt, dass die Schleife "fertig" ist, wenn das letzte Objekt verarbeitet wurde.

### ```range```-Funktion
Möchte man ganz viele aufeinanderfolgende Zahlen (nacheinander) durchgehen, kann man schlecht vorher eine lange Liste von Zahlen aufschreiben (vor allem, weil man sowas andauernd benötigt). Vor allem, wenn man sehr viele Zahlen benötigt, ergibt es Sinn, eine Funktion zu benutzen, die einem die Zahlenfolge generiert. Python liefert die Funktion ```range```, die genau das macht.

```range``` liefert keine Liste, wie wir sie von Python kennen, es handelt sich um ein eigenes Objekt. Tatsächlich ist das sogar gut so, da so (theoretisch) unendlich lange Listen aufeinanderfolgender Zahlen generiert werden können, ohne dass sie gespeichert werden müssen! In einem späteren Termin werden wir das Konzept von Generatoren und Iteratoren kennenlernen, die sehr ähnlich funktionieren.

*Achtung*: Python fängt immer mit 0 an zu zählen! ```range(4)``` generiert also nicht die Folge "1, 2, 3, 4" sondern die Folge "0, 1, 2, 3".

````python
liste = range(4)
print(liste)    # range ist ein Objekt
print(list(liste)) # Mit dem list-Befehl kann man iterierbare Objekte in Listen umwandeln

for index in liste:
    print(index)
````

### Bonus: List-Comprehension
Sehr kurze For-Schleifen lassen sich in einer Zeile durchführen! Die Idee ist, dass man so sehr schnell (unter anderem) Listen befüllen kann, ohne extra eine ganze For-Schleife programmieren zu müssen.

````python
# "Normaler" Ansatz
liste1 = []
for index in range(10):
    liste1.append(2*index-1)
print(liste1)

# List-Comprehension
liste2 = [2*index-1 for index in range(10)]
print(liste2)
````
