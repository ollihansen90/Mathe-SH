## 🧩 Aufgabe 4: Ein Raster mit GridBox erstellen

### 🔤 Was ist eine GridBox?

Eine **GridBox** ordnet Widgets in einem **Raster** an.

Ein Raster ist wie ein Spielfeld oder eine Tabelle:

⬜ ⬜ ⬜  
⬜ ⬜ ⬜  
⬜ ⬜ ⬜

---

### 🧠 Wofür braucht man eine GridBox?

Eine GridBox ist praktisch, wenn wir viele Elemente ordentlich anordnen möchten.

Zum Beispiel für:

* ein Spielfeld
* ein Dashboard
* eine Bildergalerie
* ein Menü mit vielen Buttons

---

### 🔹 Wichtige Eigenschaft

Bei einer GridBox benutzen wir `layout=widgets.Layout(...)`.

Damit können wir festlegen, wie das Raster aussehen soll.

Wichtig ist besonders:

* `grid_template_columns` → bestimmt, wie viele Spalten das Raster hat

Beispiel-Idee:

`grid_template_columns="repeat(3, 100px)"`

Das bedeutet:  
Wir möchten **3 Spalten**, jede Spalte ist **100px breit**.

---

### 🔹 Vorgehen
