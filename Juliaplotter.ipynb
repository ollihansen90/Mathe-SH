```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

height, width = 1000, 1000
xmin, xmax = -2, 2
ymin, ymax = -2, 2
kmax = 1000
c = -0.1 + 0.65j
c = -0.6 + 0.6j

# https://matplotlib.org/stable/tutorials/colors/colormaps.html
CM = cm.get_cmap('viridis') # z.B. ['viridis', 'plasma', 'inferno', 'magma', 'cividis']

def build_img(c, kmax, xmin, xmax, ymin, ymax, width=100):
    x = np.linspace(xmin, xmax, width).reshape((1, width))
    y = np.linspace(ymin, ymax, width).reshape((width, 1))

    z = x + 1j*y
    c = np.full(z.shape, c)
    div_time = np.zeros(z.shape, dtype=int)
    div = np.full(c.shape, True, dtype=bool)

    for i in range(kmax):
        z[div] = z[div]**2 + c[div]
        div[np.abs(z) > 2] = False
        div_time[div] = i

    return div_time

plt.figure(figsize=[10,10])
plt.imshow(CM(build_img(c, kmax, xmin, xmax, ymin, ymax, width=1000)))
plt.show()
```
