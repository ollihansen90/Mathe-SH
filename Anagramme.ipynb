## üìú Die Wortliste

Im n√§chsten Schritt laden wir eine gro√üe **Wortliste** aus einer Textdatei herunter.  
Diese enth√§lt fast **2 Millionen verschiedene W√∂rter**.  
Jedes Wort steht dabei in einer eigenen Zeile der Datei.

Mit folgendem Code laden wir die Datei, zerlegen sie in einzelne W√∂rter und speichern sie in der Variable `wortliste`:


```python
import requests
r = requests.get("https://gist.githubusercontent.com/MarvinJWendt/2f4f4154b8ae218600eb091a5706b5f4/raw/36b70dd6be330aa61cd4d4cdfda6234dcb0b8784/wordlist-german.txt")
wortliste = r.text.split("\n")
wortliste = sorted(list(set([w.upper() for w in wortliste])))
print("Die Liste enth√§lt",len(wortliste),"W√∂rter.")
```
